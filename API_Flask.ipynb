```python
import pandas as pd
from flask_ngrok import run_with_ngrok
from flask import request, jsonify, Flask
import random as rk
```

```python
app = Flask(__name__) #the name of the application package
run_with_ngrok(app)
```

```python
d = [
    {
        "id": 1,
        "name": "Mahesh",
        "age": 25,
        "city": "Bangalore",
        "country": "India"

    },
    {
        "id": 2,
        "name": "Alex",
        "age": 26,
        "city": "London",
        "country": "UK"

    }
]
```
